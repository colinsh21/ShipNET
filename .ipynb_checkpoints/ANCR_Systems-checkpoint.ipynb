```python
def project_current_distribution_bus(g, LA_I,bus={}):
    """
    Projects the current distributions on the logical architecture to the physical architecture
    In:
    LA_I = logical architecture with current distribution on edges, {sys:logical network: nodes+edges:prob occupied}
    g = physical architecture
    bus = bus locations for distribution system, {sys:nodes+edges}

    Out:
    g_current = physical architecture network with probability of occupancy,
        g_current[ele]={('sys1'):p_sys1, ('sys2'):p_sys2}
    """
    g_comp = g.copy()  # Track complementary probability P(not sys)

    for sys, l_net in LA_I.iteritems():
        # print sys
        # create initial complementary node and edge values
        nx.set_node_attributes(g_comp, sys, 1.0)
        nx.set_edge_attributes(g_comp, sys, 1.0)

        for i, j, d in l_net.edges(data=True):  # get edges within logical architecture
            # d={node,edge: prob occupied}
            #             print i,j,d
            for n in g_comp.nodes():
                g_comp.node[n][sys] *= (1.0 - d['i'][n])

            for ni, nj in g_comp.edges():
                # print (ni,nj), d[(ni,nj)]
                # print g_comp.edge[ni][nj][sys]
                if (ni, nj) not in d['i']:
                    g_comp[ni][nj][sys] *= (1.0 - d['i'][(nj, ni)])  # multiply by compliment
                else:
                    g_comp[ni][nj][sys] *= (1.0 - d['i'][(ni, nj)])  # multiply by compliment

                    # for n,d in g_comp.nodes(data=True):
                    # print n,d

                    # for e1,e2,d in g_comp.edges(data=True):
                    # print (e1,e2),d

    g_current = g_comp.copy()  # Get current flow: 1-Pc

    for sys in LA_I:
        if sys in bus:
            sys_bus=bus[sys]
        else:
            sys_bus=nx.Graph()
        for n in g_current.nodes():
            if n in sys_bus.nodes():
                g_current.node[n][sys]=1.0
            else:
                g_current.node[n][sys] = 1.0 - g_comp.node[n][sys]
        for i, j in g_current.edges():
            if ((i,j) in sys_bus.edges()) or ((j,i) in sys_bus.edges()):
                g_current[i][j][sys]=1.0
            else:
                g_current[i][j][sys] = 1.0 - g_comp.edge[i][j][sys]

    """
    for n,d in g_current.nodes(data=True):
        print n,d

    for e1,e2,d in g_current.edges(data=True):
        print (e1,e2),d
    """

    return g_current
```

```python
def comp_walk(g,s_comp,t_comp):
    """
    Guided walk from component s to component t through physical architecture based on current distribution
    In: g=physical architecture network, s_comp={loc:prob}, t_comp={loc:prob}
    Out: w=list of edges in walk
    """
    #Select component location
    s_locs=s_comp.keys()
    s_probs=s_comp.values()
    idx=np.random.choice(len(s_locs),1,p=s_probs)
    s=s_locs[np.squeeze(idx)]
    
    t_locs=t_comp.keys()
    t_probs=t_comp.values()
    idx=np.random.choice(len(t_locs),1,p=t_probs)
    t=t_locs[np.squeeze(idx)]
    
    w=walk_st(g,s=s,t=t)
    
    return w,s,t
```

```python
def head_loss_st(w,s,t,V=3,Di=0.05,e=1E-5,l_ratio=(10,10,10)):
    """
    Calculate head loss for a pump system routing
    In: w=walk in network space, L_ratio=conversion of edge into lengths (x,y,z), 
        pipe flow params: V=m/s, Di=inner diameter (m), e=average roughness (m), 
    Out: h_loss=head loss
    """
    rho=1000 #fluid density
    mu=1E-3 #fluid viscosity
    eD=e/Di #roughness ratio
    
    vert_distance=t[-1]-s[-1]*l_ratio[2] #z-coord difference
    
    #Get number of turns and length of the walk
    length=0.0
    turns=0.0
    dir_list=[]
    for i,j in w:
        e_dir=np.nonzero([a-b for a,b in zip(i,j)])[0][0]
#         print length
        length+=l_ratio[e_dir]
        if dir_list:
            if e_dir!=dir_list[-1]:
                turns+=1
        dir_list.append(e_dir)
#     print dir_list,turns

    #Get friction loss, K
    Re=Reynolds(V=V,D=Di,rho=rho,mu=mu)
    fd=friction_factor(Re,eD=eD)
    K = K_from_f(fd=fd, L=length, D=Di)
    K+=entrance_sharp()
    K+=exit_normal()
    K+=turns*bend_rounded(Di=Di,angle=90,fd=fd)
    h_loss=head_from_K(K,V=V)
    h_loss+=vert_distance
    
    #calcuate horsepower required
    dP_pipes=P_from_head(head=h_loss,rho=rho)
    q=V*(Di/2.0)**2*np.pi #volume in m^3/s
    p=dP_pipes*q #power in watts
    hp=p/745.7 #horse_power calc
    
    return h_loss, hp, length
    
    
    
    
```

```python
def walk_st(g,s,t):
    """
    Guided walk from node s to t through physical architecture based on current distribution
    In: g=physical architecture network, s=source node, t=sink node
    Out: w=list of edges in walk
    """
    w=[]
    loc=s
    i_dict=current_st(g,s,t)
    
    while loc != t:
        i_vals=[] #current values on edges
        
        #iterate through neigbhor edges
        nodes=g.neighbors(loc)
        edges=[]
#         print nodes
        for n in nodes:
#             print (loc,n)
            if (loc,n) in i_dict:
                e=(loc,n)
            else:
                e=(n,loc)
#             print e
            edges.append(e)
            if e not in w: #has not been used in the walk
                    i_vals.append(i_dict[e]) #can use that edges
            else: #has been used in walk
                i_vals.append(0.0) #don't use again
                
        sum_i=sum(i_vals)
        if sum_i==0:
            probs=[1.0/len(i_vals) for i in xrange(len(i_vals))] #transition probabilities are uniform
        else:
            probs=[i/sum_i for i in i_vals] #transition probabilities based on i
#         print nodes
#         print probs
        idx=np.random.choice(len(nodes),1,p=probs) #select next location based on transition probabilities
        loc=nodes[np.squeeze(idx)]
        w.append(edges[np.squeeze(idx)])
#         print w,loc
        
    return w
            
    
```

```python
def i_to_numpy(g,i):
    """
    Converts current distribution to a network
    In: g=physical architecture network, i=dictionary of current distribution {element: i}
    Out: m=numpy matrix with i weightings
    """
    edge_i={e:i[e] for e in g.edges()}
    nx.set_edge_attributes(g,'weight',edge_i)
    return nx.to_numpy_matrix(g)
    
```
